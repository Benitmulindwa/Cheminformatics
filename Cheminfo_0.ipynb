```python
from rdkit import Chem
molecule = Chem.MolFromSmiles('NC(C(C)C)C(NC(C(C)C)C(NC(C(C)C)C(O)=O)=O)=O')
unique_smiles = Chem.MolToSmiles(molecule, canonical=True,isomericSmiles=True)
print (unique_smiles)
```

```python
molecule='C(-[*:1])(=O)[H]'
r_groups = ['[H]', 'C', 'ERROR',]

molecules = [ molecule.replace('-[*:1]', r) for r in r_groups ]
print (molecules)
```

2. **Molecule** **validator**

```python
valid_molecules = []
failure_molecules = []
for molecule in molecules:
   try:
       rdkit_molecule = Chem.MolFromSmiles(molecule)
       valid_molecules.append(Chem.MolToSmiles(rdkit_molecule))
   except:
       failure_molecules.append(molecule)
print ('Valid Molecules: %s' % valid_molecules)
print ('Failure Molecules: %s' % failure_molecules)
```

# 3. **Mol2 file**

```python
class Mol2Parser(object):

    __version__ = '0.0.1'

    def __init__(self, mol2_file):

        self.mol2_file = open(mol2_file, 'r')
        self.mol2_lines = self.mol2_file.readlines()

        # Confirm

        _ = [ print(line.strip('\n')) for line in self.mol2_lines ]

if __name__ == '__main__':

    mol2_parser = Mol2Parser('benzene.mol2')
```
